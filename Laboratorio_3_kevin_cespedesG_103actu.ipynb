```python
import pandas as pd
import matplotlib.pyplot as plt

# Especifica la ruta del archivo
ruta = "/content/drive/MyDrive/conjunto de datos/atenciones_abogadas_de_las_mujeres - atenciones_abogadas_de_las_mujeres.csv"

# Lee el archivo CSV en un DataFrame
df = pd.read_csv(ruta)

# Verifica las primeras filas del DataFrame para asegurarte de que los datos sean correctos
print(df.head())

# Suponiendo que 'anio_mes' es una columna de meses representados como nmeros (1 a 12)
# Puedes ordenarlos manualmente en el orden cronolgico deseado
meses_ordenados = ['2019-03', '2019-04', '2019-05', '2019-06', '2019-07', '2019-08', '2019-09', '2019-10', '2019-11', '2019-12',
                   '2020-01', '2020-02', '2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08', '2020-09', '2020-10',
                   '2020-11', '2020-12', '2021-01', '2021-02', '2021-03', '2021-04', '2021-05', '2021-06', '2021-07', '2021-08',
                   '2021-09', '2021-10', '2021-11', '2021-12', '2022-01', '2022-02', '2022-03']

# Cuenta las ocurrencias de cada valor nico en la columna 'anio_mes'
conteo = df["anio_mes"].value_counts().reindex(meses_ordenados, fill_value=0)

# Crea un grfico de lnea
plt.figure(figsize=(10, 6))  # Establece el tamao de la figura
plt.plot(conteo.index, conteo, color="blue", marker='o', linestyle='-', markersize=5, label='Conteos')
plt.xlabel('Ao-Mes')
plt.ylabel('Conteo')
plt.title('Conteos de "anio_mes" a lo largo del tiempo (ordenados cronolgicamente)')
plt.xticks(rotation=45)  # Rota las etiquetas del eje x para una mejor legibilidad
plt.grid(True)
plt.legend()
plt.tight_layout()  # Asegura que las etiquetas y ttulos no se corten

# Muestra el grfico
plt.show()
```
